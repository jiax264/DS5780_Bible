### Merge BBE Test results from MVP & Gemma into same dataset

### Import packages

```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None) 
pd.set_option('display.max_rows', None) 
```

```python
gemma_df = pd.read_csv('gemma_test_data.csv')
mvp_df = pd.read_csv('mvp_test_data.csv')
```

```python
df = pd.merge(gemma_df, mvp_df, on=['chapter', 'KJV', 'BBE'])
```

```python
df
```

### Remove rows where Gemma refused to generate BBE text

```python
selected_rows = [11, 19, 26, 27, 34, 35, 36, 45, 65, 67, 99] # gemma refused to generate bbe for these rows 

no_gemma_bbe = df.iloc[selected_rows]

complete_test_df = df.drop(selected_rows)
```

#### Inspect these rows (look at 'gemma_bbe')

```python
no_gemma_bbe
```

```python
complete_test_df
```

### Remove rows where text is just a list of names or where KJV and BBE diverge too much

```python
selected_rows = [28, 64, 80, 41, 53, 58, 70, 84, 85, 88, 96] # rows that are just lists of names or KJV and ground truth BBE too different 

dirty_rows_df = df.iloc[selected_rows]

complete_test_df = complete_test_df.drop(selected_rows)
```

```python
dirty_rows_df
```

```python
complete_test_df.reset_index(drop=True, inplace=True)
```

```python
complete_test_df.to_csv('final_merged_test_df.csv', index=False)
```

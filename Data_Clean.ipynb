## Data Cleaning 

### import packages

```python
import os
import pandas as pd
import re
```

```python
pd.set_option('display.max_colwidth', None) 
```

### Collect each chapter from text file in KJV and BBE folders respectively and put them into a dataframe where each row is (chapter name, KJV text, BBE text)

```python
def read_file_content(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()
```

```python
kjv_base_dir = 'KJV'
bbe_base_dir = 'BBE'

data_entries = []

# Loop through each file in the KJV directory
for root, dirs, files in os.walk(kjv_base_dir):
    for file_name in files:
        if file_name.endswith('.txt'):
            # Remove the .txt extension from the file name
            chapter_name = os.path.splitext(file_name)[0]

            # Construct file paths for both KJV and BBE files
            kjv_file_path = os.path.join(root, file_name)
            bbe_file_path = kjv_file_path.replace(kjv_base_dir, bbe_base_dir)
            
            # Read the contents of the files
            kjv_content = read_file_content(kjv_file_path)
            bbe_content = read_file_content(bbe_file_path)
            
            # Append the data to the list
            data_entries.append({
                'chapter': chapter_name,
                'KJV': kjv_content,
                'BBE': bbe_content
            })

df = pd.DataFrame(data_entries)
```

#### Take a look at sample of dataframe, raw text from KJV and BBE 

```python
df.sample(3)
```

### Clean raw text by removing leading '1', and removing \n characters

```python
def process_text(text):
    # Remove the leading '1 ' from the start of the text if it exists
    if text.startswith('1 '):
        text = text[2:]  # Strip the initial '1 ' from the text

    # Replace newline followed by one or more digits and keep the space
    text = re.sub(r'\n(\d+)\s', ' ', text)
    
    # Replace standalone newline characters that are not followed by a number
    text = re.sub(r'\n', '', text)
    return text
```

```python
df_processed = df.copy()
df_processed['KJV'] = df_processed['KJV'].apply(process_text)
df_processed['BBE'] = df_processed['BBE'].apply(process_text)
```

```python
df_processed.to_csv('bible_cleaned_data.csv', index=False)
```
